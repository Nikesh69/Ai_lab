```python
import ast
```

```python
 def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python

movies.dropna(inplace=True)
```

#### It show the name only in dataset like Action,Adventure etc

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

### It show the 3 actor name only 

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['character'])
        counter+=1
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

### It sho wthe director name only

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace("","") for i in x])
```

### The code replaces empty strings with empty strings in the lists contained in the keywords, cast, crew, and overview columns of the movies DataFrame.

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace("","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace("","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace("","") for i in x])
movies['overview'] = movies['overview'].apply(lambda x:[i.replace("","") for i in x])
```

### The code concatenates and combines the text from the overview, genres, keywords, cast, and crew columns in the movies DataFrame into a single string for each row, separating the text from each column with a space.

```python
 movies['tags'] = movies['overview'].apply(lambda x: ' '.join(x)) + ' ' + \
                 movies['genres'].apply(lambda x: ' '.join(x)) + ' ' + \
                 movies['keywords'].apply(lambda x: ' '.join(x)) + ' ' + \
                 movies['cast'].apply(lambda x: ' '.join(x)) + ' ' + \
                 movies['crew'].apply(lambda x: ' '.join(x))
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
 new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x.split()))
```

```python
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem (text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

### Result
